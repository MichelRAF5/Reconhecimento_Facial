# Definição do Problema

Nas empresas, o controle de acesso de colaboradores e
visitantes ainda é feito de forma manual ou com cartões
físicos, o que gera filas, riscos de extravio e falhas de
segurança. Um sistema de reconhecimento facial pode
automatizar o processo de identificação, reduzindo erros
e aumentando a eficiência operacional.

#  Objetivo do Projeto

Desenvolver um sistema de reconhecimento facial para
controle de acesso empresarial, capaz de identificar
automaticamente funcionários e visitantes autorizados,
garantindo agilidade na entrada, rastreabilidade e
segurança das instalações.

# Tipo de Problema de IA

Trata-se de um problema de Visão Computacional, com
tarefas de Detecção e Reconhecimento Facial, em que o
sistema identifica rostos em tempo real e os compara com
um banco de dados de pessoas cadastradas.

# Dados Necessários

Imagens faciais de colaboradores e visitantes obtidas com
consentimento, capturadas em diferentes condições de
iluminação, ângulos e expressões. Esses dados formam a
base para treinamento do modelo e validação de desempenho
em ambiente real.

# Técnicas e Modelos

Utilização de Redes Neurais Convolucionais(CNNs) e ferramentas como OpenCV e DeepFace para detecção e reconhecimento
facial. Modelos baseados em embeddings faciais garantem alta precisão ao comparar rostos autorizados com os detectados
pelas câmeras.

# Métricas de Avaliação

Avaliação por Acurácia, Precisão, Revocação(Recall) e
F1-Score, medindo a taxa de acertos na identificação
correta dos indivíduos e a redução de falsos positivos e
negativos.

# Desafios e Limitações

Desafios incluem variações de iluminação, uso de
máscaras, mudança de aparência e vieses de dados de
treinamento. Limitações também envolvem custos de
implementação e atenção à privacidade, conforme
exigências da LGPD.

# Reflexão Final

O reconhecimento facial aplicado ao controle de acesso
empresarial representa um avanço tecnológico estratégico,
trazendo mais segurança, eficiência e rastreabilidade
para as organizações. Por outro lado, requer uma
abordagem ética, com políticas de consentimento e
proteção de dados bem definidas.
